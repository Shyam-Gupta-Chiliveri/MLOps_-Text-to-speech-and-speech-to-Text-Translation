## Part 1: Introduction to TTS and STT

### What is Text-to-Speech (TTS)?

Text-to-Speech (TTS) is a technology that converts written text into spoken words. It's widely used in:
- **Accessibility**: Helping visually impaired users access written content
- **Voice assistants**: Siri, Alexa, Google Assistant
- **Navigation systems**: GPS voice guidance
- **Educational tools**: Language learning applications

### What is Speech-to-Text (STT)?

Speech-to-Text (STT), also known as Automatic Speech Recognition (ASR), converts spoken language into written text. Applications include:
- **Voice assistants**: Understanding user commands
- **Transcription services**: Converting meetings, interviews to text
- **Voice typing**: Dictating emails, documents
- **Accessibility**: Helping hearing-impaired users
- **Call centers**: Automated customer service

### OpenAI's Audio APIs

OpenAI provides two main audio APIs:
1. **TTS API**: Uses advanced neural networks to generate natural-sounding speech
2. **Whisper API**: Based on OpenAI's Whisper model for highly accurate speech recognition

## Part 3: Text-to-Speech (TTS) with OpenAI

### Understanding TTS Parameters

OpenAI's TTS API offers several customization options:

- **model**: Currently "tts-1" (faster) or "tts-1-hd" (higher quality)
- **voice**: Different voice options (alloy, echo, fable, onyx, nova, shimmer)
- **input**: The text you want to convert to speech
- **response_format**: Audio format (mp3, opus, aac, flac)
- **speed**: Playback speed (0.25 to 4.0)



```python
def text_to_speech_basic(text, voice="alloy", model="tts-1"):
    """
    Convert text to speech using OpenAI's TTS API
    
    Args:
        text (str): Text to convert to speech
        voice (str): Voice to use (alloy, echo, fable, onyx, nova, shimmer)
        model (str): Model to use (tts-1 or tts-1-hd)
    
    Returns:
        bytes: Audio data
    """
    try:
        # Make API call to OpenAI TTS
        response = client.audio.speech.create(
            model=model,
            voice=voice,
            input=text,
            response_format="mp3"
        )
        
        # Return the audio content
        return response.content
        
    except Exception as e:
        print(f"Error in TTS: {e}")
        return None

```

```python
# Example usage
sample_text = "Hello! Welcome to our Text-to-Speech lesson using OpenAI's API."

# Generate speech
audio_data = text_to_speech_basic(sample_text)

if audio_data:
    # Save to file
    with open("sample_output.mp3", "wb") as f:
        f.write(audio_data)
    
    # Display audio player in Jupyter
    display(Audio(audio_data, autoplay=False))
    print("Audio generated successfully!")
else:
    print("Failed to generate audio")
```

### Advanced TTS Features


```python
def advanced_text_to_speech(text, voice="alloy", model="tts-1-hd", speed=1.0):
    """
    Advanced TTS with more options
    
    Args:
        text (str): Text to convert
        voice (str): Voice selection
        model (str): Model quality (tts-1 or tts-1-hd)
        speed (float): Speech speed (0.25 to 4.0)
    
    Returns:
        bytes: Audio data
    """
    try:
        response = client.audio.speech.create(
            model=model,
            voice=voice,
            input=text,
            response_format="mp3",
            speed=speed
        )
        
        return response.content
        
    except Exception as e:
        print(f"Error in advanced TTS: {e}")
        return None
```

```python
# Example with different speeds
long_text = """
Artificial Intelligence is transforming the way we interact with technology. 
Text-to-speech systems like OpenAI's TTS API make it possible to create 
natural-sounding voice applications with just a few lines of code.
"""

speeds = [0.5, 1.0, 1.5, 2.0]

for speed in speeds:
    print(f"Speed: {speed}x")
    audio_data = advanced_text_to_speech(long_text, speed=speed)
    
    if audio_data:
        display(Audio(audio_data, autoplay=False))
    print("-" * 30)
```

## Part 4: Speech-to-Text (STT) with Whisper

### Understanding Whisper API

OpenAI's Whisper API provides:
- **High accuracy**: State-of-the-art speech recognition
- **Multiple languages**: Supports 99+ languages
- **Transcription**: Convert audio to text
- **Translation**: Convert non-English audio to English text
- **Format flexibility**: Supports various audio formats

### Basic STT Example

```python
def speech_to_text_basic(audio_file_path):
    """
    Convert speech to text using OpenAI's Whisper API
    
    Args:
        audio_file_path (str): Path to audio file
    
    Returns:
        str: Transcribed text
    """
    try:
        # Open the audio file
        with open(audio_file_path, "rb") as audio_file:
            # Make API call to Whisper
            transcript = client.audio.transcriptions.create(
                model="whisper-1",
                file=audio_file,
                response_format="text"
            )
        
        return transcript
        
    except Exception as e:
        print(f"Error in STT: {e}")
        return None
```

```python

# First, let's create an audio file to transcribe using our TTS
text_to_transcribe = "This is a test sentence for speech recognition using OpenAI's Whisper API."

# Generate audio
audio_data = text_to_speech_basic(text_to_transcribe)

if audio_data:
    # Save audio file
    test_audio_file = "test_for_stt.mp3"
    with open(test_audio_file, "wb") as f:
        f.write(audio_data)
    
    print("Original text:", text_to_transcribe)
    print("Generated audio file:", test_audio_file)
    
    # Now transcribe it back to text
    transcribed_text = speech_to_text_basic(test_audio_file)
    
    if transcribed_text:
        print("Transcribed text:", transcribed_text)
        print("Match:", text_to_transcribe.lower() == transcribed_text.lower().strip())
    else:
        print("Transcription failed")
```

### Advanced STT with Detailed Response


```python

def speech_to_text_detailed(audio_file_path, language=None):
    """
    Speech to text with detailed response format
    
    Args:
        audio_file_path (str): Path to audio file
        language (str): Language code (optional, e.g., 'en', 'es', 'fr')
    
    Returns:
        dict: Detailed transcription result
    """
    try:
        with open(audio_file_path, "rb") as audio_file:
            # Request detailed JSON response
            transcript = client.audio.transcriptions.create(
                model="whisper-1",
                file=audio_file,
                response_format="verbose_json",
                language=language  # Optional: specify language
            )
        
        return transcript
        
    except Exception as e:
        print(f"Error in detailed STT: {e}")
        return None

```

```python
# Example with detailed response
detailed_result = speech_to_text_detailed("test_for_stt.mp3")

if detailed_result:
    print("Detailed Transcription Results:")
    print(f"Text: {detailed_result.text}")
    print(f"Language: {detailed_result.language}")
    print(f"Duration: {detailed_result.duration:.2f} seconds")
    
    # Print word-level timestamps if available
    if hasattr(detailed_result, 'words') and detailed_result.words:
        print("\nWord-level timestamps:")
        for word in detailed_result.words[:10]:  # Show first 10 words
            print(f"  {word.word}: {word.start:.2f}s - {word.end:.2f}s")
```

### Translation Feature


```python

def speech_to_english_translation(audio_file_path):
    """
    Translate non-English speech directly to English text
    
    Args:
        audio_file_path (str): Path to audio file
    
    Returns:
        str: Translated English text
    """
    try:
        with open(audio_file_path, "rb") as audio_file:
            # Use translations endpoint instead of transcriptions
            translation = client.audio.translations.create(
                model="whisper-1",
                file=audio_file,
                response_format="text"
            )
        
        return translation
        
    except Exception as e:
        print(f"Error in translation: {e}")
        return None

```

```python
# Example: Create french audio and translate to English
french_text = " Excusez-moi, o se trouve la gare?"

# Generate french audio
french_audio = text_to_speech_basic(french_text, voice="nova")

if french_audio:
    french_audio_file = "french_test.mp3"
    with open(french_audio_file, "wb") as f:
        f.write(french_audio)
    
    print("Original french text:", french_text)
    display(Audio(french_audio, autoplay=False))
    # Transcribe (should detect french)
    french_transcription = speech_to_text_basic(french_audio_file)
    print("French transcription:", french_transcription)
    
    # Translate to English
    english_translation = speech_to_english_translation(french_audio_file)
    print("English translation:", english_translation)
    
```

### Whispaer + GPT

```python
from openai import OpenAI

client = OpenAI()  

# Step 1: Transcribe the audio
def transcribe_audio(audio_path):
    with open(audio_path, "rb") as audio_file:
        transcription = client.audio.transcriptions.create(
            model="whisper-1",
            file=audio_file,
            response_format="text",
            language='en'
        )
    return transcription

# Step 2: Translate using GPT
def translate_text(text, target_language="Spanish"):
    prompt = f"Translate the following to {target_language}:\n\n{text}"
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.3
    )
    return response.choices[0].message.content.strip()
```

```python
french_text = transcribe_audio("nationalite1.mp3")
german_translation = translate_text(french_text, target_language="German")
print("French text:", french_text)
print("German translation:", german_translation)
```
